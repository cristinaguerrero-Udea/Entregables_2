```python
# =====================================================
# NOTEBOOK 01 - EXPLORACIÓN DE DATOS
# Proyecto: Análisis de precios de medicamentos en Colombia
# Fuente: Datos Abiertos Colombia
# =====================================================

# Importación de librerías
import pandas as pd
import numpy as np

# -----------------------------------------------------
# 1. Carga de datos desde JSON (Datos Abiertos Colombia)
# -----------------------------------------------------

url = "https://www.datos.gov.co/resource/3t73-n4q9.json"

# Cargar el dataset
df = pd.read_json(url)

# -----------------------------------------------------
# 2. Vista general del DataFrame
# -----------------------------------------------------

print("Primeras filas del dataset:")
display(df.head())

print("\nDimensiones del DataFrame (filas, columnas):")
print(df.shape)

# -----------------------------------------------------
# 3. Información general del dataset
# -----------------------------------------------------

print("\nInformación general del DataFrame:")
df.info()

# -----------------------------------------------------
# 4. Revisión de valores nulos
# -----------------------------------------------------

print("\nCantidad de valores nulos por columna:")
print(df.isnull().sum())

# -----------------------------------------------------
# 5. Estadísticas descriptivas iniciales
# (solo columnas numéricas)
# -----------------------------------------------------

print("\nEstadísticas descriptivas:")
display(df.describe())

# -----------------------------------------------------
# 6. Listado de columnas disponibles
# -----------------------------------------------------

print("\nColumnas del dataset:")
for col in df.columns:
    print("-", col)

# -----------------------------------------------------
# 7. Observaciones iniciales
# -----------------------------------------------------

print("\nObservaciones iniciales:")
print("- El dataset contiene información sobre precios de medicamentos.")
print("- Se identifican valores nulos que deberán tratarse en la fase de limpieza.")
print("- Algunas columnas requerirán conversión a formato numérico.")
print("- Se realizará limpieza y análisis estadístico en los siguientes notebooks.")
```
