```python
# =====================================================
# NOTEBOOK 03 - ANÁLISIS Y VISUALIZACIONES
# Proyecto: Análisis de precios de medicamentos en Colombia
# Fuente: Datos Abiertos Colombia
# =====================================================

# Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración estética
sns.set(style="whitegrid")

# -----------------------------------------------------
# 1. Carga y preparación básica de los datos
# -----------------------------------------------------

url = "https://www.datos.gov.co/resource/3t73-n4q9.json"
df = pd.read_json(url)

# Normalizar columnas
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Conversión de precio a numérico
df['precio_por_tableta'] = pd.to_numeric(df['precio_por_tableta'], errors='coerce')

# Eliminación de nulos
df = df.dropna(subset=['precio_por_tableta'])

# Eliminación de outliers (IQR)
Q1 = df['precio_por_tableta'].quantile(0.25)
Q3 = df['precio_por_tableta'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['precio_por_tableta'] >= Q1 - 1.5 * IQR) &
        (df['precio_por_tableta'] <= Q3 + 1.5 * IQR)]

# -----------------------------------------------------
# 2. Análisis estadístico descriptivo
# -----------------------------------------------------

media = df['precio_por_tableta'].mean()
mediana = df['precio_por_tableta'].median()
moda = df['precio_por_tableta'].mode()[0]

varianza = df['precio_por_tableta'].var()
desviacion = df['precio_por_tableta'].std()
rango = df['precio_por_tableta'].max() - df['precio_por_tableta'].min()

print("ANÁLISIS ESTADÍSTICO DE PRECIOS")
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda:.2f}")
print(f"Varianza: {varianza:.2f}")
print(f"Desviación estándar: {desviacion:.2f}")
print(f"Rango: {rango:.2f}")

# -----------------------------------------------------
# 3. Tabla de frecuencias (medicamentos más frecuentes)
# -----------------------------------------------------

if 'nombre_medicamento' in df.columns:
    frecuencia = df['nombre_medicamento'].value_counts().head(10)
    print("\nTop 10 medicamentos más frecuentes:")
    print(frecuencia)

# -----------------------------------------------------
# 4. Visualizaciones
# -----------------------------------------------------

# 4.1 Histograma
plt.figure(figsize=(8,5))
plt.hist(df['precio_por_tableta'], bins=30)
plt.title("Distribución de precios de medicamentos")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()

# 4.2 Boxplot
plt.figure(figsize=(6,4))
plt.boxplot(df['precio_por_tableta'])
plt.title("Boxplot de precios de medicamentos")
plt.ylabel("Precio")
plt.show()

# 4.3 Gráfico de barras (Top 10 precios promedio)
if 'nombre_medicamento' in df.columns:
    top_precios = df.groupby('nombre_medicamento')['precio_por_tableta'].mean().sort_values(ascending=False).head(10)

    plt.figure(figsize=(10,5))
    top_precios.plot(kind='bar')
    plt.title("Top 10 medicamentos con mayor precio promedio")
    plt.xlabel("Medicamento")
    plt.ylabel("Precio promedio")
    plt.xticks(rotation=45, ha='right')
    plt.show()

# 4.4 Heatmap de correlaciones
plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="coolwarm")
plt.title("Mapa de calor de correlaciones")
plt.show()

# -----------------------------------------------------
# 5. Interpretación general
# -----------------------------------------------------

print("\nInterpretación general:")
print("- Los precios presentan una distribución asimétrica.")
print("- Existen medicamentos con precios significativamente más altos.")
print("- La dispersión indica variabilidad importante en el mercado farmacéutico.")
print("- Los resultados pueden apoyar análisis sobre acceso y regulación de medicamentos.")
```
